**Retrieve Money Data**

```
with open('/content/drive/My Drive/Colab Notebooks/data/cc.csv') as csvfile:
    read = csv.reader(csvfile, delimiter=',')
    Ncc = {}
    Mcc = {}
    i = -1
    for row in read:
      if i == -1: i = 0;
      else:
        x = int(row[0])
        if x in Ncc:
          count = Ncc[x]+1
          Ncc[x] = count
        else:
          Ncc[x] = 1
        if x in Mcc:
          amount = Mcc[x]+float(row[2])
          Mcc[x] = amount
        else:
          Mcc[x] = float(row[2])
```

```
with open('/content/drive/My Drive/Colab Notebooks/data/kplus.csv') as csvfile:
    read = csv.reader(csvfile, delimiter=',')
    Nkp = {}
    Mkp = {}
    i = -1
    for row in read:
      if i == -1: i = 0;
      else:
        x = int(row[0])
        if x in Nkp:
          count = Nkp[x]+int(row[2])
          Nkp[x] = count
        else:
          Nkp[x] = int(row[2])
        if x in Mkp:
          amount = Mkp[x]+float(row[3])
          Mkp[x] = amount
        else:
          Mkp[x] = float(row[3])
print(Nkp)
print(Mkp)
```

**Data Prep for DL**

```
tx = []
ty = [[i] for i in train]
for i in range(50000):
  tx.append([])
  tx[i].append((int(gender[i+1])+1)%2)
  tx[i].append(int(ocp[i+1]))
  tx[i].append(int(age[i+1]))
  Nsum, Msum = 0,0
  Nsum2, Msum2 = 0,0
  for j in card[i+1]:
    if j in Ncc:
      Nsum = Nsum + Ncc[j]
    if j in Mcc:
      Msum = Msum + Mcc[j]
    if j in Nkp:
      Nsum2 = Nsum2 + Nkp[j]
    if j in Mkp:
      Msum2 = Msum2 + Mkp[j]
  tx[i].append(Nsum)
  tx[i].append(Msum)
  tx[i].append(Nsum2)
  tx[i].append(Msum2)
  tx[i].append(Nsum+Nsum2)
  
  tx[i].append(Msum+Msum2)
  if (Nsum+Nsum2) != 0:
    tx[i].append((Msum)/(Nsum+1))
  else:
    tx[i].append(0.0)
```

**Train Model**

```
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0, 1))
trainx = sc.fit_transform(tx)
trainy = sc.fit_transform(ty)
X = array(trainx)
X = X.reshape((X.shape[0], X.shape[1], 1))
trainy = [i[0] for i in trainy]
y = trainy
```

```
print(X.shape)
```

```
# Import Keras backend
import keras.backend as K
# Define SMAPE loss function
def smape(true,predicted):
    epsilon = 0.1
    summ = K.maximum(K.abs(true) + K.abs(predicted) + epsilon, 0.5 + epsilon)
    smape = K.abs(predicted - true) / summ * 2.0
    return smape
import keras.losses
keras.losses.smape = smape
import keras.metrics
keras.metrics.smape = smape
```

```
from keras.models import Sequential,load_model
from keras.layers import LSTM,Conv1D
from keras.layers import Dense,Dropout,MaxPooling1D,Flatten
from keras.callbacks import EarlyStopping, ModelCheckpoint

model = Sequential()

#model.add(Conv1D(filters=100, kernel_size=7, activation='relu', input_shape=(7,1)))
#model.add(Dropout(0.1))
#model.add(Conv1D(filters=100, kernel_size=1, activation='relu'))
#model.add(Dropout(0.1))
"""
model.add(Conv1D(filters=100, kernel_size=1, activation='relu'))
model.add(Dropout(0.1))
model.add(Conv1D(filters=100, kernel_size=1, activation='relu'))
model.add(Dropout(0.1))
model.add(Conv1D(filters=100, kernel_size=1, activation='relu'))
model.add(MaxPooling1D(pool_size=1))
model.add(Conv1D(filters=100, kernel_size=1, activation='relu'))
model.add(Dropout(0.1))
model.add(Conv1D(filters=100, kernel_size=1, activation='relu'))
model.add(Dropout(0.1))
model.add(Conv1D(filters=100, kernel_size=1, activation='relu'))
model.add(Dropout(0.1))
model.add(Conv1D(filters=100, kernel_size=1, activation='relu'))

#model.add(Flatten())
#model.add(Dense(100, activation='relu'))
#model.add(Dense(1))
"""
model.add(Conv1D(filters=10, kernel_size=1, activation='relu', input_shape=(10, 1)))
model.add(LSTM(units = 50))
#model.add(Dropout(0.2))
"""
model.add(LSTM(units = 50, return_sequences = True))
#model.add(Dropout(0.2))

model.add(LSTM(units = 50, return_sequences = True))
#model.add(Dropout(0.2))

model.add(LSTM(50))
#model.add(Dropout(0.2))
"""
model.add(Dense(1))
```

```
start = time.time()
#model.compile(optimizer='adam', loss='mse')

model.compile(loss=smape, optimizer="adam", metrics=[smape])
# fit model
history = model.fit(X, y, epochs=300, verbose=1,validation_split=0.2, callbacks=[EarlyStopping(monitor='val_loss',
                                                                                                     min_delta=0,
                                                                                                     patience=30,
                                                                                                     verbose=1,
                                                                                                     mode='min'),ModelCheckpoint('/content/drive/My Drive/Colab Notebooks/model/newbest_model9_11try1.h5', monitor='val_loss', mode='min', save_best_only=True)])

print("> Compilation Time : ", time.time() - start)
```
